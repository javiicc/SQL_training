If set A = {L M N O P} and the set B = {P Q R S T}, what sets are generated by the following operations?

- A union B

A ∪ B = {L M N O P Q R S T}

- A union all B

- A intersect B

## Exercise 6-2

Write a compound query that finds the first and last names of all actors and customers whose last name starts with L.

```python
pd.read_sql_query(
    "SELECT a.first_name, a.last_name\
       FROM actor a\
       WHERE a.last_name LIKE 'L%'\
     UNION \
     SELECT c.first_name, c.last_name\
       FROM customer c\
       WHERE c.last_name LIKE 'L%'", 
    conn)
```

## Exercise 6-3

Sort the results from Exercise 6-2 by the `last_name` column.

```python
pd.read_sql_query(
    "SELECT a.first_name fname, a.last_name lname\
       FROM actor a\
       WHERE a.last_name LIKE 'L%'\
     UNION \
     SELECT c.first_name, c.last_name\
       FROM customer c\
       WHERE c.last_name LIKE 'L%'\
     ORDER BY lname", 
    conn)
```
