## Exercise 3-1

Retrieve the actor ID, first name, and last name for all actors. Sort by the last name and then by first name. 

```python
pd.read_sql_query(
    'SELECT actor_id, first_name, last_name\
      FROM actor\
      ORDER BY last_name, first_name\
      LIMIT 10', 
    conn
)
```

## Exercise 3-2

Retrieve the actor ID, first name, and last name for all actors whose last name equals `WILLIAMS` or `DAVIS`.

```python
pd.read_sql_query(
    "SELECT actor_id, first_name, last_name\
      FROM actor\
      WHERE last_name = 'WILLIAMS' OR last_name = 'DAVIS'", 
    conn
)
```

## Exercise 3-3

Write a query againts the `rental` table that returns the IDs of the customers who rented a film on July 5, 2005.

```python
pd.read_sql_query(
    "SELECT customer_id\
      FROM rental\
      WHERE date(rental_date) = '2005-07-05'", 
    conn
)
```
